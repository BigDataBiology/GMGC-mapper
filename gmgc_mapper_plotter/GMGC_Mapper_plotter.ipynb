```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("genome_bin.tsv", header=1)
df= df.iloc[:,0].str.split('\t', expand=True)
```

```python
hist_2= df[1].value_counts(ascending=True)
k_2 = len(df[1])
graf_2 = hist_2.plot(kind='barh', figsize=(10, 10), color='powderblue', zorder=5, width=0.5)
graf_2.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")
vals_2 = graf_2.get_xticks()
for tick in vals_2:
      graf_2.axvline(x=tick, linestyle='dashed', alpha=0.8, color='lightgrey', zorder=1)
graf_2.set_xlabel("Counts", labelpad=10, weight='bold', size=12)
graf_2.set_ylabel("Number of hits", labelpad=10, weight='bold', size=12)
graf_2.set_title("GMGC mapper", weight='bold', size=14)
for i, v in enumerate(hist_2):
    graf_2.text(v + 2 , i - 0.2, str(v)+'('+str(round(v*100/k_2 , 1))+'%)', color='teal', fontweight='bold',size=8)
```

```python
hist_2= df[1].value_counts()
k_2 = len(df[1])
graf_3 = hist_2.plot(kind='barh', figsize=(10, 10), color='powderblue', zorder=5, width=0.5)
graf_3.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")
vals_3 = graf_2.get_xticks()
for tick in vals_3:
      graf_3.axvline(x=tick, linestyle='dashed', alpha=0.8, color='lightgrey', zorder=1)
graf_3.set_xlabel("Counts", labelpad=10, weight='bold', size=12)
graf_3.set_ylabel("Number of hits", labelpad=10, weight='bold', size=12)
graf_3.set_title("GMGC mapper", weight='bold', size=14)
for i, v in enumerate(hist_2):
    graf_3.text(v + 2 , i - 0.2, str(v)+'('+str(round(v*100/k_2 , 1))+'%)', color='teal', fontweight='bold',size=8)
```

```python
x = ['<10 hits', '10 - 25 hits', '25 - 50 hits', '>50 hits' ]

k , j , l , m = 0 , 0 , 0 , 0
for i in df[1]:
    if int(i) < 10:
        k += 1
    if 10<= int(i) < 25:
        j += 1
    if 25 <= int(i) < 50:
        l += 1
    if int(i) >= 50:
        m += 1
y = [k,j,l,m]
t = y[0] + y[1]+ y[2] + y[3]
```

```python
graf_4 = plt.figure(figsize=(8, 6))
ax = graf_4.gca()
ax.bar(x,y, color='powderblue')
#ax.set_xlabel("", labelpad=10, weight='bold', size=12)
ax.set_ylabel("Counts", labelpad=10, weight='bold', size=12)
ax.set_title("GMGC mapper", weight='bold', size=14)
for i  in (range(len(y))):
    ax.text(i , y[i]+3 , str(y[i])+'('+str(round(int(y[i])*100/int(t),1))+'%)', horizontalalignment='center', color='teal', fontweight='bold')
print()
```
